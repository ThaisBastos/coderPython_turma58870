```python
#exercicio 1 
#Crie um array NumPy com 10 elementos inteiros aleatorios no intervalo [0,100] e exiba os valores e o tipo de dado do array

import numpy as np

#Escreve valores aleatorios
array_aleatorio = np.random.randint(0, 101, size=10)

print("Valores do array:", array_aleatorio)
print("Tipo de dado do array:", array_aleatorio.dtype)
```

```python
#exercicio 2 - matriz de 5x5 com valores aleatorios de [0,1)
#Crie uma matriz de 5x5 com valores aleatorios de [0,1) e encontre o valor minimo, máximo e a media da matriz

import numpy as np

#Criacao da matriz
matriz_numpy = np.random.rand(5, 5)

#Calculos
minimo_numpy = np.min(matriz_numpy)
maximo_numpy = np.max(matriz_numpy)
media_numpy = np.mean(matriz_numpy)

print("Matriz NumPy:")
print(matriz_numpy)
print(f"Mínimo: {minimo_numpy}")
print(f"Máximo: {maximo_numpy}")
print(f"Média: {media_numpy}")
```

```python
#exercicio 3 - Criar um array com 10 elementos aleatorio no intervalo [0,1]
#Crie um array com 10 elementos aleatorio no intervalo [0,1]. Em seguida, crie um novo array com os elementos do primeiro array com os elementos do primeiro array multiplicados por 10 e convertidos para inteiro.
import numpy as np

# Criar um array com valores aleatorios
array_original = np.random.rand(10)

# Criar um novo array 
array_novo = (array_original * 10).astype(int)


print("Array Original:")
print(array_original)
print("\nArray Novo:")
print(array_novo)
```

```python
#exercicio 4 - array de duas dimensões com shape (3,3)
#Crie um array de duas dimensões com shape (3,3) e valores aleatórios inteiros no intervalo [0,9]. Substitua todos os elementos da segunda linha por -1.
import numpy as np
import pandas as pd

array = np.random.randint(0, 10, size=(3, 3))

array[1, :] = -1

print("Array de duas dimensões:")
print(array)
```
